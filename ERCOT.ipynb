```python
ercot = pd.read_excel('ERCOTNorth.xlsx' , sheet_name= None , parse_dates=[['Delivery Date', 'Hour Ending']] )
```

```python
ercot
```

```python
full_table = pd.DataFrame()
for name, sheet in ercot.items():
    sheet['sheet'] = name
    sheet = sheet.rename(columns=lambda x: x.split('\n')[-1])
    full_table = full_table.append(sheet)

full_table.reset_index(inplace=True, drop=True)
```

```python
full_table= full_table[['Delivery Date_Hour Ending','Settlement Point', 'Settlement Point Price']]
```

```python
full_table.set_index('Delivery Date_Hour Ending', inplace= True)
```
